```python
#Medium1
def dictionary(s):
    d={}
    for i in s:
        c=i.find(":")
        d[i[0:c]]=int(i[c+1:])
    return d
a=input().split(',')
a=dictionary(a)
b=input().split(',')
b=dictionary(b)
d={}
for i in a:
    if i in b:
            d[i]=a[i]+b[i]
    else:
        d[i]=a[i]
for i in b:
    if i not in a:
            d[i]=b[i]
print(d)
v=d.values()
l=[]
for i in v:
    if i not in l:
        l.append(int(i))
for i in range(0,len(l)-1):
    for j in range(i+1,len(l)):
        if l[i]>l[j]:
            temp=l[i]
            l[i]=l[j]
            l[j]=temp
l=tuple(l)
print(l)
    
```

```python
#Hard1 altanative
d={input().upper():input().upper()}
for x,y in d.items():
    c=1
    for i in x:
        if x.count(i)!=y.count(i):
            c=0
            break
    if c==1:
         print('Those strings are anagrams.')
    else:
         print('Those strings are not anagrams.')
        
        
```
