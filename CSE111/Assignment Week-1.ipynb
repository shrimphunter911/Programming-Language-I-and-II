```python
#Easy-4
n=int(input("Enter a value"))
d=0
for i in range(1,n,1):
    if n%i==0:
        d+=1
if d==1:
    print("Prime")
else:
    print("Not prime")
```

```python
#Hard-2
n=int(input("Enter an integer"))
for r in range(1,n+1):
    for c in range(1,2*n):
        if r==n or r+c==n+1 or c-r==n-1:
            print("*", end="")
        else:
            print(end=" ")
    print()
for j in range(1,n):
    for k in range(1,2*n):
        if k-j==1 or k+j==2*n-1:
            print("*", end="")
        else:
            print(end=" ")
    print()
```
